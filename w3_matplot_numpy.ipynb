### Visualization of iris sepal w n len

```python
f= open("iris.txt")

allines=f.readlines()

x_values=[]
y_values=[]
species_to_colors= {}
color_codes=[]

for line in allines[1:]:
    line=line.strip()
    info=line.split(",")
    sepal_len=float(info[1])
    sepal_w=float(info[2])
    species=info[-1]
    
    if species not in species_to_colors:
        species_to_colors[species]= len(species_to_colors) # our len of dict, it will be 1-2-3 to each specie will get different no
      
    color_codes.append(species_to_colors[species])  #0-1-2-3...
    

    x_values.append(sepal_len)
    y_values.append(sepal_w)
    
    
    
f.close()
print(color_codes)
```

```python
plt.scatter(x_values,y_values, c=color_codes)
plt.xlabel("Sepal Length in cm")
plt.ylabel("Sepal width in cm")
plt.show()
```

#### Now lets show the legends so we can see the species

```python
f= open("iris.txt")

allines=f.readlines()

d={}

for line in allines[1:]:
    line=line.strip()
    info=line.split(",")
    sepal_len=float(info[1])
    sepal_w=float(info[2])
    species=info[-1]
    
    if species not in d:
        d[species]= [[sepal_len],[sepal_w]] #list in list
        
    else:
        d[species][0].append(sepal_len)
        d[species][1].append(sepal_w)
        
    
    
f.close()

print(d)

```

```python
for k,v in d.items():
    plt.scatter(v[0], v[1], label=k)
    
plt.legend()
plt.xlabel("Sepal Length in cm")
plt.ylabel("Sepal width in cm")
plt.show()
```

## Google Stock Prices every weekday since 2004

```python
f = open("GOOGL.txt")
all_lines = f.readlines()

d = {}

for line in all_lines[1:]:
    line = line.strip()
    info = line.split(",")
    date = info[0]
    year = int(date.split("-")[0])
    
    if 2004 < year < 2022:  # we do not want 2022 and 2004 since they are not complete
        close = float(info[4])
        if year not in d:
            d[year] = {"days": [], "stocks": []}  # dict in dict
            
        d[year]["days"].append(len(d[year]["days"]) + 1)  # +1 since no 0 day
        d[year]["stocks"].append(close)

f.close()

for k, v in d.items():
    plt.plot(v["days"], v["stocks"], label=str(k))

plt.legend()
plt.figure(figsize=(20,10)) #size of the plot

plt.show()
```
